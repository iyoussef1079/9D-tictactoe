```python
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def spherical_coordinate(x, y):
    return [math.cos(x) * math.cos(y), math.sin(x) * math.cos(y), math.sin(y)]

def NX(n, x):
    pts = []
    start = -1.0 + 1.0 / (n - 1)
    increment = (2.0 - 2.0 / (n - 1.0)) / (n - 1.0)
    for j in range(0, n):
        s = start + j * increment
        pts.append(spherical_coordinate(s * x, math.pi / 2.0 * math.copysign(1, s) * (1.0 - math.sqrt(1.0 - abs(s)))))
    return pts

def generate_points(n):
    return NX(n, 0.1 + 1.2 * n)

def plot_points_with_sphere(points):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    xs = [p[0] for p in points]
    ys = [p[1] for p in points]
    zs = [p[2] for p in points]

    ax.scatter(xs, ys, zs, color='r', label='Points')

    # Draw sphere
    u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
    x = np.cos(u) * np.sin(v)
    y = np.sin(u) * np.sin(v)
    z = np.cos(v)
    ax.plot_wireframe(x, y, z, color="b", alpha=0.1)

    # Setting limits for better visualization
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])
    ax.set_zlim([-1, 1])
    ax.set_xlabel('X Coordinate')
    ax.set_ylabel('Y Coordinate')
    ax.set_zlabel('Z Coordinate')
    ax.set_title('9 Equidistant Points on a Sphere with Spherical Outline')
    ax.legend()

    plt.show()

# Generate and plot points
n_points = 9
points = generate_points(n_points)
print("Points on the sphere: ", points)
plot_points_with_sphere(points)
```
